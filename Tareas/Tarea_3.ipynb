se utiliza para instalar la biblioteca de Python llamada Geocoder. Esta herramienta permite realizar operaciones de geocodificación, lo que implica convertir direcciones físicas (como direcciones postales) en coordenadas geográficas (latitud y longitud), y viceversa.


```python
!pip install geocoder
```

```python
import pandas as pd
import geocoder

# Función para estandarizar direcciones utilizando la biblioteca geocoder
def normalizar_direccion(direccion):

    # Comprobar si la dirección es nula o está vacía
    if pd.isna(direccion):
        return None  # Puedes optar por devolver un valor vacío o cualquier otro valor predeterminado

    # Realizar una limpieza básica de la dirección, eliminando espacios extra y convirtiendo a minúsculas
    direccion = direccion.lower().strip()

    # Utilizar geocoder para obtener una versión estandarizada de la dirección a través de ArcGIS
    g = geocoder.arcgis(direccion)

    # Comprobar si la geocodificación ha encontrado una dirección válida
    if g.ok:
        # Devolver la dirección formateada o estandarizada
        return g.address
    else:
        return None  # O bien devolver la dirección original o un mensaje de error si no se encuentra

# Aplicar la función de estandarización a la columna 'direccion' del DataFrame
df['direccion_normalizada'] = df['direccion'].apply(normalizar_direccion)
df['direccion'] = df['direccion'].apply(normalizar_direccion)
df
```

```python
# Eliminar todas las filas que tienen un valor None en la columna 'direccion_normalizada'
df = df.dropna(subset=['direccion_normalizada'])
df
```

```python
# Reemplazar todos los valores NR y NaN en las columnas por 'Sin respuesta'
df = df.fillna('Sin respuesta')
df
```

```python
# Normalizar las columnas 'material' y 'transformacion' eliminando tildes y caracteres especiales
df['material'] = df['material'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df['transformacion'] = df['transformacion'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df
```

```python
# Se borra la columna de dirección
df = df.drop('direccion', axis=1)
df
```

Clasificar los materiales en un DataFrame en diferentes categorías, como "Plástico", "Papel/Cartón", "Vidrio", "Metal", "Orgánico" y "Neumáticos", basándose en palabras clave específicas. Para cada valor de la columna de materiales, la función identifica si contiene alguna de las palabras clave asociadas a una categoría y asigna la categoría correspondiente. Si no encuentra una coincidencia, se asigna "Sin respuesta". El resultado es una nueva columna que contiene la categoría asignada a cada material.

```python
def categorizar_material(materiales):
    """Clasifica los materiales según sus características y similitudes.

    Args:
        materiales: Un texto que describe los materiales.

    Returns:
        Una lista de categorías correspondientes a los materiales.
    """
    categorias = []
    materiales = str(materiales).lower()  # Convertir a minúsculas para facilitar la comparación

    # Verificar si el material pertenece a alguna categoría relacionada con plásticos
    if any(palabra in materiales for palabra in ["plastico", "plastico flexible", "plastico rigido", "plastico rigido y flexible", "polipropileno", "poliestireno"]):
        categorias.append("Plástico")

    # Verificar si el material pertenece a la categoría de papel o cartón
    if any(palabra in materiales for palabra in ["papel", "carton", "plegadisas", "papel periodico", "cartulina", "kraft", "cajas de carton"]):
        categorias.append("Papel/Cartón")

    # Verificar si el material es de vidrio
    if any(palabra in materiales for palabra in ["vidrio", "envases", "botellas"]):
        categorias.append("Vidrio")

    # Verificar si el material es metálico
    if any(palabra in materiales for palabra in ["metal", "aluminio", "acero", "latas", "hierro", "cobre", "chatarra"]):
        categorias.append("Metal")

    # Verificar si el material es orgánico
    if any(palabra in materiales for palabra in ["organico", "biodegradable", "compostable", "residuos", "alimentos", "restos de comida"]):
        categorias.append("Orgánico")

    # Verificar si el material pertenece a la categoría de neumáticos
    if any(palabra in materiales for palabra in ["neumaticos", "llantas", "caucho"]):
        categorias.append("Neumáticos")

    # Si no se encontró ninguna categoría, se asigna "Sin respuesta"
    if not categorias:
        categorias.append("Sin respuesta")

    return categorias

# Aplicar la función de categorización a la columna 'material'
df['categoria_material'] = df['material'].apply(categorizar_material)
df
```
